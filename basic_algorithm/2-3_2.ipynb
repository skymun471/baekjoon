```python
# 가장 많이 등장하는 문자열을 출력하세요.(다른 풀이)
text = input("문자열을 입력하시오: ")
text_list = []
text_dic = {}
for i in text:
  if  i not in text_list:
    text_list.append(i)
    text_dic.update({i:1})
  else:
    text_dic[i] += 1 

print(text_dic)
pre_val = 0
for k in text_dic:
  print(k)
  if text_dic[k] > pre_val:
    big_text = k
    pre_val = text_dic[k]
print(f"most_text is '{big_text}'")
```

## 자리 바꾸기 swap
나중에 나올 버블정렬에서 기초적으로 쓰이니 잘 알아두기

```python
# 입력한 두 자리 숫자의 자리를 바꾼 결과를 출력하는 함수를 만들어 보세요.
# 파이썬 함수의 매개변수 type 지정 방법 : val_name : type
# 
def swap(a):
  first = a // 10
  second = a - (first * 10)
  val = None
  val = first 
  first = second
  second = val
  print(first*10 + second)

swap(35)
swap(89)
```

## 짝수 홀수 구하는 함수 만들기

```python
# 짝수면 true 홀수면 false 가 나오는 함수를 만드시오.
# even 짝수
# odd 홀수

def is_odd(a):
  if 0 != (a%2):
    return True
  else:
    return False
print(is_odd(34))
```

```python
#  == 로 연산된 식은 Ture False 값을 return 한다.
def is_even_2(a):
  result = a % 2 == 0
  return result
is_even_2(4)
```

## 배수 인지 알아보기

```python
# 배수(multiple)란 몇 배를 한 수로 10은 2의 배수 이고 11은 2의 배수가 아니다 2 * 5는 10이지만 11은 표현 할 수 없다. 
# 0은 모든 수의 배수이다.
# 어떤 수 의 배수 라는 의미는 그 수를 어떤 수로 나누었을 떄 나머지가 0인 수 이다.
```

```python
# %연산자로 나먼지 구하기
#  두 숫자 a, b 가 주어졌을 때 a 가 b의 배수이면 Ture 아니면 False 를 리턴하는 함수를 구하시오.
def multiple(a, b):
  result = a%b == 0
  return result

multiple(0,3)
```

```python
# 자릿수 들의 합구하기
# 숫자의 각 자릿수를 모두 더한 값을 리턴 하는 함수를 만들어보시오.
# // 연산 수행시 몫만을 구할 수 있고 % 수행시 나머지를 구할 수 있습니다. 이를 이용해서 10을 이용한 몫 나머지 연산을 통해 각자리의 숫자를 구할 수 있습니다.

def all_sum(a):
  #  몫: quotient, 나머지: remainder
  result = 0
  while(a>0):
    q = a // 10
    r = a % 10
    result += r
    a = q

  return result + q

all_sum(1234) 
```

```python
# 최댓값 max 최솟값 min 구하기
# 다음 리스트 안에 최댓값을 구하는 함수를 만드시오.

list_a = [9,22,3,33,4,5]
def max(a):
  result = 0
  print(a)
  for i in range(len(a)):
    if result < a[i]:
      result = a[i]
  return result

max(list_a)
```

```python
# 최댓값이 들어있는 인덱스를 반환하는 함수를 만드시오.
list_a = [9,22,3,33,4,5]
def max_index(list_a):
  result = 0
  idx = 0
  for i in range(len(list_a)):
    if result < list_a[i]:
      result = list_a[i]
      idx = i
  
  return result, idx

max, idx = max_index(list_a)
print('max',max,'idx',idx)
```

```python
# 다음 리스트 일때의 최댓값을 반환하는 함수를 만드시오.
# max 함수와 max_index 함수는 result의 초기값을 0으로 설정 해줬기 때문에 음수보다 작은 값을 list에 포함하고 있다면 제대로된 계산이 이루어지지 않은다.
# 때문에 초기값을 list 값의 처음값으로 지정해줬다.
# 위 함수와 다른 점이라면 len()을 사용하지 않고 구현했다는 것이다. 그렇다면 len()은 어떤 상황에서 사용하는 것일까?
# len()을 이용하면 해당 리스값의 인덱스 값을 
list_b = [-9,-22,-3,-33,-4,-5]

def max_b(list):
  result = list[0]
  for i in list:
    if result < i:
      result = i
  return result

max_b(list_b)
```

```python
# 불필요한 연산 제거하기
list_b = [-9,-22,-3,-33,-4,-5]

def max_c(list):
  result = list[0]
  for i in list[1:]:
    if result < i:
      result = i
  return result

max_c(list_b)
```

```python
# 최적화하기
list_b = [-9,-22,-3,-33,-4,-5]
#  index 값을 통해서 값을 불러와 중복되지 않게 비교하는 것이 최적화에 도움이 된다.
def max_d(list):
  result_idx = 0
  for i in range(1, len(list)):
    if list[i] > list[result_idx]:
      result_idx  = i
  
  return list[result_idx]

max_d(list_b)
```
